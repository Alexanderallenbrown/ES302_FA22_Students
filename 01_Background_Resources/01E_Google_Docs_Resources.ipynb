# Google Docs for ES302

You have likely used Google Docs for assignments before. In ES302, we will use Google Docs for your process writing assignments. All of these except the first one will require you to write technical documents in the [IEEE style](https://www.ieee.org/content/dam/ieee-org/ieee/web/org/conferences/style_references_manual.pdf). Therefore, each of you has been given access to a template that you can copy and move into your drive to use as a starting point for your assignments. The template is available [here](https://docs.google.com/document/d/1BMLExxD-IPdg8BpoX1QPa5_iaZHdTEeQ5yxBzSZsEWo/edit?usp=sharing). You can make a copy of this document and move it using the Google Docs interface.

You are expected to place all assignments in the folder I've shared with you called ES302_YourLastName. You can find this folder by searching for it under "Shared With Me" on the Google Drive web interface at www.drive.google.com


Once you find this folder, you can feel free to add a shortcut to it in your Drive so you can find it easily. You can do this by right clicking on the folder, and selecting "Add Shortcut To Drive." This will also allow you to access the folder from the google drive [desktop client](https://www.google.com/drive/download/), which will be important for using this folder to complete your robot assignments (which is also required). 

# Citations in Google Drive with Zotero

The technical writing you will be doing this semester will require you to cite references in the IEEE style. This can get tedious, especially if you have many references (and you will!). 

## Getting Zotero

To make things easier, make an account with the reference manager [Zotero](www.zotero.org) and download the Zotero Connector for Chrome **and** the desktop application. 


When downloading the Connector extension for Chrome, be sure to install and enable it appropriately.


# Using Zotero Connector to Save References

After Zotero Connector and Zotero Desktop are installed, let's say you're browsing for a good article on fish-robot interaction. You happen to find one! With the Zotero Connector, it only takes a couple of clicks to add this paper to your library so that you can cite it in a paper. 


Once you've done this, you should be able to see the reference in your Zotero Library on the Zotero Desktop App.


You should also set up your local Zotero library to sync with the Zotero cloud so that your references are available anywhere! This can be done in the preferences menu of Zotero.


# Citing References From your Zotero Library in Google Docs

Now that you have all of your favorite sources saved in your Zotero library, you can also use the Zotero Connector to easily cite sources and create a bibliography in Google Docs. A full tutorial is available [here](https://www.zotero.org/support/google_docs). For ES302's writing assignments, you will need to ensure that your citations and bibliography are in IEEE format.

# Equations in Google Docs

To easily typeset equations in Google Docs, we will be using the LaTeX language implemented by the Google Docs Extension "Auto-LaTeX Equations." To install this extension, go to "Extensions->Add Ons->Get Add ones" from a Google Docs document.
